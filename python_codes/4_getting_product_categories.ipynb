## WEB SCRAPING MULTIPLE PAGES

```python
# empty list
product_category = []
index = []

# iteration for webscrapping multiple pages
for link, i in zip(all12.link, range(8317, len(all))):
    r = requests.get(f"https://www.smallable.com{link}")
    soup = BeautifulSoup(r.content, 'html.parser')
    print(f'Response for {link} :', r.status_code)
    #scraping infos from site
    #product category
    category = soup.find('nav', attrs={"class": "BreadCrumb_breadcrumb__3eArq BreadCrumb_breadcrumbLastItem__3KOwJ"}).select('span')[-1].get_text(strip=True)
    product_category.append(category)
    #index
    index.append(i)
    wait_time = random.randint(2,5)
    print("I will sleep for " + str(wait_time) + " seconds.")
    sleep(wait_time) #Script will be stopped for a period of time ranging from 2 to 6 seconds
```
