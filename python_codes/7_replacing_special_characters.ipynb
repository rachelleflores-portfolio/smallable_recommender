# Replacing Special Characters
- This step has been done in order to load the csv in SQL Workbench. Unfortunately, it is not able to read the file with accented letters.

## IMPORT LIBRARIES

```python
import pandas as pd
```

```python
all = pd.read_csv(r'./../data/cleaned_data/filtered_all.csv')
```

```python
def change_char(df):
    """Function to replace special characters with normal alphabet"""
    special_characters = ["[àâä]","[éèêë]", "ï", "[öòóô]", "[ùûü]", "[²]", "[’]", "[°]"]
    replace = ["a", "e", "i", "o", "u", "2", "'", ""]

    for s, r in zip(special_characters, replace):
        for column in df.columns:
            df[column] = df[column].replace(s, r, regex=True)
    
    return df
        
```

```python
all_new = change_char(all)
```

```python
# save with encoding ascii required by sql and ignore any other errors from other special characters
all_new.to_csv("sql_csv.csv", index=False, encoding="ascii", errors="ignore")
```
